# This notebook is for analyzing the Adverse Food Events data from Kaggle

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('adverse-food-events/CAERS_ASCII_2004_2017Q2.csv')
```

```python
df.head(-5)
```

```python
df.describe(include='all')
```

```python
# find what data is missing
print('the number of missing age values are: {}'.format(np.isnan(df['CI_Age at Adverse Event']).sum()))
df['Age_Filtered'] = df['CI_Age at Adverse Event'][df['CI_Age at Adverse Event']<100]
```

```python
df['Age_Filtered'].describe()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
sns.countplot(df['Age_Filtered'])
```

```python
sns.countplot(df['Age_Filtered'][~np.isnan(df['Age_Filtered'])])
```

```python
outcomes = df['AEC_One Row Outcomes'].str.get_dummies(sep=', ')
outcomes.describe()
```

```python
outcome_data = outcomes.sum().sort_values(ascending=False)
print(outcome_data)
# plt.xticks(rotation=90)
```

```python
symptoms = df['SYM_One Row Coded Symptoms'].str.get_dummies(sep=', ')
symptoms.describe()
```

```python
symptom_data = symptoms.sum().sort_values(ascending=False)
print(symptom_data[:5])
```

```python
female_temp = df['Age_Filtered'][df['CI_Gender']=='Female']
female_age_temp = female_temp[~np.isnan(df['Age_Filtered'])]
```

```python
fig, ax = plt.subplots(figsize=(20,8))
sns.countplot(female_age_temp)
```

```python
male_temp = df['Age_Filtered'][df['CI_Gender']=='Male']
male_age_temp = male_temp[~np.isnan(df['Age_Filtered'])]
fig, ax = plt.subplots(figsize=(20,8))
sns.countplot(male_age_temp)
```

```python
df = df.drop(columns = ['CI_Age at Adverse Event'])
```

```python
# fig, ax = plt.subplots(figsize=(20,8))
sns.countplot(df['PRI_FDA Industry Name'],order=df['PRI_FDA Industry Name'].value_counts().iloc[:5].index)
plt.xticks(rotation=75)
plt.title('count of incidents by product type')
```

```python
print('the number of men who report incidents with vitamins is {}'.format(df['PRI_FDA Industry Name'][(df['CI_Gender']=='Male') & (df['PRI_FDA Industry Name']=='Vit/Min/Prot/Unconv Diet(Human/Animal)')].count()))
print('the number of women who report incidents with vitamins is {}'.format(df['PRI_FDA Industry Name'][(df['CI_Gender']=='Female') & (df['PRI_FDA Industry Name']=='Vit/Min/Prot/Unconv Diet(Human/Animal)')].count()))
```
