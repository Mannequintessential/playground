## Analysis of the Wine Dataset from Kaggle
We are going to go through the wine dataset from Kaggle. The goal is to explore the dataset look for unique insights


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
# import the data
df = pd.read_csv('winemag-data_first150k.csv')
```

```python
# get the first 5 rows to get a picture of what dataframe looks like
df.head()
```

## Data Cleaning

```python
# use describe to get a high level understanding of each column. use include='all' 
# to get stats on text based columns as well
df.describe(include='all')
```

This is a good way to find NaNs in the dataset. If you look at the count row, not all of them add up to 150930. This lets us know where we need to do some data cleaning and fix those empty values. Right off the bat, we see that country is missing 5 entries, designation is missing almost 50k entries, price has about 13k missing values. 

We need to figure out what the best way to deal with this is. In the case of country, we can remove the 5 missing values without much of an impact. For the other variables, we will investigate further to see what the best approach is to handle it. 

Variables with missing values:
* country > listwise deletion
* designation 
* price
* province
* region_1
* region_2


```python
# digging into the designation data

# this will give us the number of unique designations for each wine entry, along with the count of each. 
print(df['designation'].value_counts())

print('the number of unique designations is {}'.format(len(df['designation'].value_counts())))
```

There are 30621 different designations. But wait, we see that there are two entries for Barrel Sample. It looks like they use different cases for sample so it is treated as uniqe. Let's run this again, but first convert the entries to lower case. 

```python

# this will give us the number of unique designations for each wine entry, along with the count of each. 
df['designation'] = df['designation'].str.lower()
print(df['designation'].value_counts())

print('the number of unique designations is {}'.format(len(df['designation'].str.lower().value_counts())))
```

The number of unique designations went from 30625 to 30435. That's almost 200 designations that were consolidated. Not bad! 

Next we notice that reserve, reserva, and riserva show up in the top five entries. It looks like they all carry the "reserve" designation, but in different languages. As we have a country variable already, having separate designations for separate languages doesn't give us any new information. We'll replace various translations of reserve into English. We'll also combine 'gran reserve' and 'grand reserve'. 

```python
df.loc[df.designation.isin(['reserva','riserva','réserve'])] = 'reserve'
df.loc[df.designation.isin(['gran reserva'])] = 'grand reserve'
print(df['designation'].value_counts())
```

Now we see some designations like 'estate reserve' and 'special reserve'. Or 'estate grown' and 'estate bottled'. these bigrams have separate meanings so we'll keep them separate for now. 

Now we need to deal with the missing values. In this case, missing likely means there is no special designation. It doesn't mean that the entry is containg an error. So rather than doing an imputation of mode, we'll just create a 'none' designation to say there is no unique additor to the wine name. 

```python
df.loc[df.designation.isin(['NaN'])] = 'none'
```

```python
# prepare models
models = []
# models.append(('LR', LogisticRegression()))
# models.append(('LDA', LinearDiscriminantAnalysis()))
# models.append(('KNN', KNeighborsClassifier()))
# models.append(('CART', DecisionTreeClassifier()))
# models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
```

```python
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
seed = 42
for name, model in models:
	kfold = model_selection.KFold(n_splits=2, random_state=seed)
	cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
```

```python
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
